# Redrob Hackathon — AI Recruiter Ranker Sandbox

This notebook runs the offline candidate ranker on a small sample. No API keys, no GPU, no external dependencies.

**To use:**
1. Run cell 1 to clone the repo and install nothing (stdlib only)
2. Upload your candidates file (JSONL format, any number of candidates) in cell 2
3. Run cell 3 to rank and see results

```python
# Cell 1: Clone repo
import subprocess
result = subprocess.run(
    ['git', 'clone', 'https://github.com/SAURABH22KAR/ai-recruiter-ranker', 'ranker'],
    capture_output=True, text=True
)
print(result.stdout or result.stderr)
import sys, os
os.chdir('/content/ranker')
print('Ready. rank.py requires no pip installs.')
```

```python
# Cell 2: Upload candidates file OR use the bundled sample
import json, os

# Option A: Use bundled sample_candidates.json (50 candidates)
sample_path = 'sample_candidates.json'
with open(sample_path) as f:
    sample = json.load(f)

# Write as JSONL for rank.py
with open('/tmp/sample.jsonl', 'w') as f:
    for c in sample:
        f.write(json.dumps(c) + '\n')

print(f'Loaded {len(sample)} candidates from sample_candidates.json')
print('(To use your own file: upload a .jsonl file and set sample_path below)')
```

```python
# Cell 3: Run the ranker
import subprocess, csv

result = subprocess.run(
    ['python', 'rank.py', '--candidates', '/tmp/sample.jsonl',
     '--out', '/tmp/output.csv', '--top-n', '10'],
    capture_output=True, text=True
)
print(result.stdout[-3000:] if len(result.stdout) > 3000 else result.stdout)
if result.stderr:
    print('STDERR:', result.stderr[-500:])
```

```python
# Cell 4: Display results as a table
import csv

with open('/tmp/output.csv') as f:
    rows = list(csv.DictReader(f))

print(f'{'Rank':<6} {'Score':<8} {'Candidate ID':<16} Reasoning')
print('-' * 100)
for row in rows:
    print(f"{row['rank']:<6} {row['score']:<8} {row['candidate_id']:<16} {row['reasoning'][:80]}")
```

```python
# Cell 5: Validate the output format
result = subprocess.run(
    ['python', 'validate_submission.py', '/tmp/output.csv'],
    capture_output=True, text=True
)
print(result.stdout or result.stderr)
```
